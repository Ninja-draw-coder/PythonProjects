```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SampleSuperstore.csv')
print(df.head())
# define data values
x = df['City'].head(10)
y = df['Sales'].head(10)

# Create a line chart
plt.figure(figsize=(10, 15))
plt.xlabel("X-axis City")
plt.ylabel("Y-axis Sales")
plt.title("Sales over city")
plt.plot(x, marker='o', linestyle='-',color='red')
plt.plot(y, marker='s', linestyle='-',color='green')




# Display grid
plt.grid(True)


# Show the plot
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('SampleSuperstore.csv')
print(df.head())
# define data values
x = df['City'].head(20)
y = df['Sales'].head(20)

# Create a line chart
plt.figure(figsize=(8, 6))

plt.bar(x,y,width=0.25,color='blue')
plt.title('Sales over city')
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```
