```python
#Step-2 Request library used to send the HTTP requestto websites and fetch their content


import requests

#Send a get request to the website

url ="https://www.ebooks.com/en-in/?srsltid=AfmBOopQ06R0wK_XuDl7gAMZ-2iHcr2yJbyar57XF3yyoKCl4wl6vode"
response = requests.get(url)

#Check if the request was successful

if response.status_code == 200:
  print("Request Successful")
  print(response.text[:1000])

else:
  print('Failed to retrieve the page')
```

```python
#Beautifulsoup -> parse and extract usefuldata from the HTML content recievd via requess

from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text,'html.parser')

#Pretiffy the html content

print(soup.prettify())
```

```python
#Extracting specific data using beautiful soup

'''Types of objects in BeautifulSoup
-> HTML tags -> <h1>,<p>
-> Navigable Strings -> text inside a tag
-> BeautifulSoup Object -> entire HTML document
-> Comments -> HTML comments  -> <!--comment-->



<html>
<head> --</head>
<title> </title>
<body>
--------------  <h1> -> heading tag  -> <h1>EBOOKS SITE</h1>
                <p> -> paragraph tag -> <p> </p>
</body>
</html>
'''
```

```python
#Extract book links

book_links = soup.find_all('a',class_ ='title')

for i , link in enumerate(book_links,start=1):
  print(f'{i}. {link.text}')
```

```python
#Extract book links

book_links = soup.find_all('a',class_ ='title')

for i , link in enumerate(book_links,start=1):
  print(f"{i}. {link['href']}")
```
