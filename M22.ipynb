```python
#MLE

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

data = np.array([101.2,100.9,101.5,100.7,101.0])

mu_values = np.linspace(100.5,101.5,100)

std = 1

#Calculate the likelihood

likelihood = [np.prod(norm.pdf(data,loc=mu,scale =std)) for mu in mu_values]

best_mu = mu_values[np.argmax(likelihood)]

print(best_mu)

#plotting
plt.plot(mu_values,likelihood,label ='Likelihood')
plt.axvline(x = best_mu,color='red',linestyle='--',label = 'Best mu')
plt.xlabel('mu')
plt.ylabel('Likelihood')
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.stats as st

data = [101.2,100.9,101.5,100.7,101.0]

#Sample mean
mean = np.mean(data)

#Sample size
n = len(data)

#Sample std
std = np.std(data,ddof=1)

#Confidence level
confidence = 0.95

#Z-score for 95% confidence
z = st.norm.ppf(1-(1-confidence)/2)

#MoE
moe = z*(std/np.sqrt(n))

#Confidence interval
lower = mean - moe
upper = mean + moe


print("Sample mean is ",round(mean,2))
print('Margin of error is ',round(moe,2))
print(f"The 95% confidence interval is ({lower:.2f},{upper:.2f})")
```
