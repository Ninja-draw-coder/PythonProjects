```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

starbucks = pd.read_csv('/content/starbucks_drinkMenu_expanded.csv')
starbucksDF = pd.DataFrame(starbucks)

#find out which “Beverage_category” has the highest calories; plot the insights using a bar plot with the help of the Seaborn library.
plt.figure(figsize=(14, 8))
plt.title('Beverage_category the highest calories')
countbyType = starbucksDF.groupby('Beverage_category').max('Calories')
countbyType = countbyType.sort_values('Calories', ascending=False)
#print(countbyType)
yaxis=countbyType.index
xaxis=countbyType['Calories']

# Plot a simple histogram and kde
sns.barplot(countbyType, x=xaxis,y=yaxis, color="m")
plt.show()


```
