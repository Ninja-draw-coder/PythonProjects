```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

# Simulated data
actual = np.array([100, 200, 300, 400, 500])
pred_rmse = np.array([90, 210, 310, 380, 490])   # Small error
pred_mape = np.array([70, 140, 270, 420, 600])   # Larger proportional error

# Calculate RMSE and MAPE
rmse = np.sqrt(mean_squared_error(actual, pred_rmse))
mape = mean_absolute_percentage_error(actual, pred_mape) * 100

# Plotting actual vs predicted
plt.figure(figsize=(12, 6))

# RMSE Plot
plt.subplot(1, 2, 1)
plt.plot(actual, label="Actual", marker='o')
plt.plot(pred_rmse, label="Predicted (for RMSE)", marker='x')
for i in range(len(actual)):
    plt.plot([i, i], [actual[i], pred_rmse[i]], 'r--')
plt.title(f"RMSE Illustration (RMSE = {rmse:.2f})")
plt.xlabel("Observation")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

# MAPE Plot
plt.subplot(1, 2, 2)
plt.plot(actual, label="Actual", marker='o')
plt.plot(pred_mape, label="Predicted (for MAPE)", marker='x')
for i in range(len(actual)):
    plt.plot([i, i], [actual[i], pred_mape[i]], 'g--')
plt.title(f"MAPE Illustration (MAPE = {mape:.2f}%)")
plt.xlabel("Observation")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_log_error

# Simulated data
actual = np.array([100, 200, 300, 400, 500])
pred_rmsle = np.array([90, 220, 280, 410, 480])  # Slight deviations

# Calculate RMSLE
rmsle = np.sqrt(mean_squared_log_error(actual, pred_rmsle))

# Plotting RMSLE visualization
plt.figure(figsize=(6, 4))
plt.plot(actual, label="Actual", marker='o')
plt.plot(pred_rmsle, label="Predicted (for RMSLE)", marker='x')
for i in range(len(actual)):
    plt.plot([i, i], [actual[i], pred_rmsle[i]], 'b--')
plt.title(f"RMSLE Illustration (RMSLE = {rmsle:.4f})")
plt.xlabel("Observation")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
