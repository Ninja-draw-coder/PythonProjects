```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Set a random seed for reproducibity
np.random.seed(42)

#Population of students scores
population = np.random.normal(loc=70,scale=10,size =1000)  #loc =mean, scale = std

#Lets take 5 samples each of size 30 from this population
sample_size = 30
num_samples = 5
samples = [np.random.choice(population,size = sample_size, replace=False) for _ in range(num_samples)]

#Calculate the mean of each sample
sample_means = [np.mean(sample) for sample in samples]

#Calculate the std of the population
population_std = np.std(population)

#Calculate standard error
standard_error = np.std(sample_means)

#Plotting
plt.figure(figsize=(10,6))
sns.histplot(sample_means,bins=5,kde=True,color='skyblue',edgecolor='black')
plt.axvline(np.mean(sample_means),color='red',linestyle='--',label=f"Mean of sample means:{np.mean(sample_means):.2f}")
plt.title("Sampling distribution of sample means")
plt.xlabel("Sample mean")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()

```
