```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

starbucks = pd.read_csv('/content/starbucks_drinkMenu_expanded.csv')
starbucksDF = pd.DataFrame(starbucks)

plt.figure(figsize=(8, 6))
plt.title('Beverage_category the highest calories')
countbyType = starbucksDF.groupby('Beverage_category').max('Calories')
df = pd.DataFrame([countbyType['Calories'], countbyType.index]).T
df = df.sort_values(by='Calories', ascending=False)
#print(df.columns[0], df.columns[1])
print(df.head(10))
x=countbyType.index
y=countbyType['Calories']

# Plot a simple histogram and kde
sns.histplot(df, kde=True, color="g",x=df.index,y=df.Calories)
plt.show()
# plt.bar(x,y)
# plt.xlabel('Beverage_category')
# plt.ylabel('Calories')
# plt.show()

```
