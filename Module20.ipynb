```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(loc = 50, scale = 10, size = 1000)
#mean=50, std=10, generate 1000 values


#Plotting
plt.figure(figsize = (10,5))
plt.hist(data,bins=30, density = True,alpha=0.6,color = 'skyblue',label ="Histogram")
xmax,xmin = plt.xlim()  #current limits of x-axis
x = np.linspace(xmin,xmax,100)
p = stats.norm.pdf(x,50,10)  #probability density function
plt.plot(x,p,'r',label="Normal Curve")
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.legend()
plt.grid()
plt.show()
```

```python
z_scores = (data-np.mean(data)) / np.std(data)

plt.figure(figsize=(10,5))
plt.hist(z_scores,bins=30,density=True,alpha=0.6,color='orange',label='Z-Histogram')

x = np.linspace(-4,4,100)
mean = np.mean(data)
std = np.std(data)
p = stats.norm.pdf(x,0,1)
plt.plot(x,p,'r',label='Z-Normal Curve')

plt.title("Z-Distribution")
plt.xlabel("Z-Scores")
plt.ylabel("Probability")
plt.legend()
plt.grid()
plt.show()
```
