```python
#Nested Loops
'''we use them when we need to repeat a set of actions multiple times within
other repeated action. It helps in scenarios like:
1. Working with tables or grids
2. Processing data in multiple layers
3. Handling nested structures
4. Performing complex calculations in structured manner.


inner loop  runs completely for each iteration of the outer loop

for loop inside for loop
for loop inside while loop
while loop inside while loop
while loop under for loop '''
```

```python
# for under for

#Multiplication table for 1 to 5

for i in range(1,6):   #outer loop for numbers 1 to 5
  print(f"\nMultiplication Table for {i}")   #f-string formatting
  for j in range(1,11):
    print(f"{i} x {j} = {i*j}")
```

```python
# for under while

#seating arrangment in bus
#-> a bus has 3 rows of seats, and each row has 4 seats. we want to assign seat numbers

row = 1  #initializer

while row<=3: #Outer loop for rows
  print(f"\nRow {row} : ")
  for seat in range(1,5):
    print(f"Seat {seat}",end=" ")  #end works like tab

  row += 1  # row = row + 1
```

```python
# while under while

#Simple Digital clock
#-> A digital clock displays hours and minutes. We want to simulate a clock from 00:00 to 2:59

hour = 0

while hour<3:
  minute = 0
  while minute<60:
    second = 0
    while second<60:
      print(f"{hour:02d}:{minute:02d}:{second:02d}") #Print time in HH:MM:SS format
      second += 1
    minute += 1
  hour += 1
```

```python
# while under for

#Find odd numbers in multiplication tables
#-> Instead of printing all multiplication table values, we only print the odd results from table 1 to 5

for i in range(1,6):
  print(f"\nOdd numbers in table of {i} : ")
  j = 1
  while j<=10:
    result = i*j
    if result % 2 !=0:
      print(f"{i} x {j} = {result}")
    j += 1
```
