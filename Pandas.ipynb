```python
'''Pandas Data Structure
->Series  = A one dimensional labeled array
->DataFrame = A 2D tabular structure, similar to a table in a database or an Excel spreadsheet

Series->
-> Homogenous
-> Immutable
-> Supports Labeled indexing
'''

#Creating a series  from a list of numpy array

import pandas as pd
import numpy as np

data = [10,20,30,40,50.9]
series1 = pd.Series(data)
print(series1)
```

```python
#Creating a series with custom index

data = [100,200,300,400]
index_labels =[100,'b','c','d']

series2 = pd.Series(data,index=index_labels)
print(series2)
```

```python
#Create a series from a dictionary

data = {'a':100,'b':200,'c':600}
series3 = pd.Series(data)
print(series3)
```

```python
#Cretae a series with scalar value

series4 = pd.Series(5,index=[1,2,3,4,5,6,7,8,9])
print(series4)
```

```python
#With numpy array

arr = np.array([10,20,30,40,50])

series5 = pd.Series(arr)
print(series5)
```

```python
'''DataFrame

can hold data of multiple types
felxible row and column indexing
supports powerful data manipulation functions like filtering , grouping and merging'''

#From list

data = [['John',28],['Alice',24],['Bob',25]]

df1 = pd.DataFrame(data,columns=['Name','Age'])
print(df1)
```

```python
df = pd.read_csv('player.csv')
```

```python
#update a cells value  in slide set_value is given which is depricated means no longer the part of pandas so we are going to use .at[]

df.at[6,'name'] = "Kwame Sitou"
df
```

```python
#filtering

#specific value with one condition

fdf = df[df['runs'] >= 5000]

fdf
```

```python
#multiple conditions

fdf = df[(df['runs']>5000) & (df['innings']>200)]

fdf
```

```python
fdf = df[(df['runs']>5000) | (df['innings']>200)]

fdf
```

```python
#isin()

fdf = df[df['100s'].isin([1,4,5])]
fdf
```
